```python
# Original dataset taken from: https://www.kaggle.com/mkechinov/ecommerce-events-history-in-cosmetics-shop?select=2019-Dec.csv
# Original dataset sttributable to: https://rees46.com/

import pandas as pd

DATA_PATH = "/Users/claytoncohn/Dropbox/New/DePaul/DSC478/FinalProject/"
```

```python
import numpy as np

df = pd.read_csv("cosmetics_data.csv", header=0, names=["event_type", "product_id", "category_id", "brand", "price", "user_id"])
print(df.shape)
df.head(10)
```

```python
# For item-based collaborative filtering, we will get a list of users with all of the items put in cart

# First we will limit the transactions only to purchases (cart)
df = df[df['event_type'] == "cart"]
print(df.shape)
df.head(10)
```

```python
# Factorize users
user_labels, _ = pd.factorize(df["user_id"])
df["user_id"] = user_labels;
df.head(25)
```

```python
# Factorize products
product_labels, _ = pd.factorize(df["product_id"])
df["product_id"] = product_labels;
df.head(25)
```

```python
df = df.drop(columns=['event_type','category_id', 'brand', "price"])
df.head(10)
```

```python
# Create dictionary of all users and transactions

users = {}
for index, row in df.iterrows():
    user = row["user_id"]
    product = row["product_id"]
    if user in users:
        products = users[user]
        products.append(product)
        users.update({ user : products })
    else:
        users.update({ user : [product] })
```

```python
# Code snippet for printing first n items of dictionary taken from:
# https://stackoverflow.com/questions/7971618/python-return-first-n-keyvalue-pairs-from-dict

from itertools import islice

def take(n, iterable):
    return list(islice(iterable, n))

head = take(10, users.items())

for row in head:
    print(row)
```

```python
# Create the array of users and procuts

import numpy as np

data = []
for key, value in users.items():
    user = key
    products = np.array(value)
    prod_unique = np.unique(products)
    
    for i in prod_unique:
        data.append([user, i, np.sum(products==i)])
```

```python
data[:10]
```

```python
df_data = pd.DataFrame(data=data, index=None, columns=["user_id", "product_id", "count"]) 
df_data.sample(10)
```
